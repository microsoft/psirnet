```python
import os
import numpy as np
from tqdm import tqdm
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import normalized_root_mse as nrmse
```

```python
files = sorted(os.listdir('../results/npz_files'))
```

```python
ssims_one_average = []
psnrs_one_average = []
nrmses_one_average = []
ssims_ours = []
psnrs_ours = []
nrmses_ours = []
for file in tqdm(files):
    data = np.load(os.path.join('../results/npz_files', file), allow_pickle=True)
    stacked = data['stacked']
    one_average = stacked[:, 0:1, ...]
    target = stacked[:, 1:2, ...]
    ours = stacked[:, 2:, ...]
    B = stacked.shape[0]  # batch size
    mn, mx = target.min(), target.max()
    target = (target - mn) / (mx - mn)  # Normalize target
    one_average = (one_average - mn) / (mx - mn)  # Normalize one_average
    ours = (ours - mn) / (mx - mn)  # Normalize
    for i in range(B):
        ssims_one_average.append(ssim(one_average[i], target[i], win_size=7, data_range=1, channel_axis=0))
        psnrs_one_average.append(psnr(one_average[i], target[i], data_range=1))
        nrmses_one_average.append(nrmse(one_average[i], target[i]))
        ssims_ours.append(ssim(ours[i], target[i], win_size=7, data_range=1, channel_axis=0))
        psnrs_ours.append(psnr(ours[i], target[i], data_range=1))
        nrmses_ours.append(nrmse(ours[i], target[i]))
ssims_one_average = 100 * np.array(ssims_one_average)
psnrs_one_average = np.array(psnrs_one_average)
nrmses_one_average = np.array(nrmses_one_average)
ssims_ours = 100 * np.array(ssims_ours)
psnrs_ours = np.array(psnrs_ours)
nrmses_ours = np.array(nrmses_ours)
```

```python
len(ssims_one_average)
```

```python
print(np.mean(ssims_one_average), np.mean(psnrs_one_average), np.mean(nrmses_one_average))
print(np.mean(ssims_ours), np.mean(psnrs_ours), np.mean(nrmses_ours))
```

```python
n = len(ssims_one_average)
print(np.std(ssims_one_average) / np.sqrt(n), np.std(psnrs_one_average) / np.sqrt(n), np.std(nrmses_one_average) / np.sqrt(n))
print(np.std(ssims_ours) / np.sqrt(n), np.std(psnrs_ours) / np.sqrt(n), np.std(nrmses_ours) / np.sqrt(n))
```
